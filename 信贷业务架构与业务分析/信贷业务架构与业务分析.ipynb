# 一、风险领域概览

## 1.金融风控算法对标的企业


1.利用互联网技术 进行金融运作的巨头企业例如蚂蚁金服，京东金融、oppo金融等，这类公司有自己的金融科技部门，有自己强大的技术团队，是
风控算法工程师趋之若鹜的dream companies；

2.非持牌网络借贷平台：例如玖富万卡等，不过因为政府对于p2p的打击，现在这类模式的平台越来越少了，剩余的也开始转换业务方向

3.持牌金融机构：和p2p网贷不同，这类公司持有正规的牌照，稳定性较好，例如拍拍贷（现更名为信也科技）、马上消费金融等

4.互联网科技公司，比较知名的例如猛犸、Datavisor以提供风控行业的解决方案——以产品或者人力或者二者兼有的方式，俗称乙方，
这类公司一般没有自己的信贷业务，主要是以产品或者人力等形式为银行等大型机构提供相应的算法服务；

5.数据商，比如典型的百融，同盾，当然这类公司也常常有一些针对风控的场景设计的一些产品；

6.银行、大型金融机构的科技部门，相对来说，技术层面上这类机构的实力偏弱，但是业务方向的沉淀是比较深厚的

1、银行体系的，例如建设银行小微快贷、个人贷，微众银行的微粒贷等；

2、持牌消金；中邮消费金融、兴业消费金融、马上消费金融等

3、小额贷款公司，百度有钱花、oppo金融、携程金融等；

4、P2P 网贷系，比如宜人贷借款、麦子金服（貌似已倒）。

目前比较常见的三种信贷产品的形态是：

1、信用卡；

2、现金贷；

3、消费贷


1、信用卡不必多说，相信基本上大部分人都了解银行的信用卡业务，一般是由银行发行、循环授信（即还即用）、申请门槛相对较高、提额麻烦的要死；


2、现金贷：现金分期产品可以看作现金贷平台的低门槛信用卡，使用过网贷的同学对此应该比较熟悉，门槛低，申请简单，额度一般不大，年化利率一般比较吓人，这两年遭遇国家大力打击，大家平常在网上看到的XXX因网贷跳楼大都是这类信贷产品。。。
现金分期应该算是最“深入人心”的网贷产品了，从互联网大厂体系的阿里的蚂蚁借呗、京东的京东金条、腾讯的微粒贷到银行系的各类百花齐放的现金贷产品再到专业海尔金融、苏宁金融最后到直接面向C端的各类小型信贷平台比如ppd、你我贷等产品


3、消费贷：消费贷产品主要以专项贷款为特点，例如针对于汽车的汽车贷款产品、专修贷款产品、旅游贷款等等；


# 二、风控业务详解


## 1.业务流程

### 业务分析

获客阶段：产品流程，渠道管理等

贷前管理：准入原则，定价策略，授信评分，反欺诈

贷中管理：交易策略，调额策略，调价策略，行为监控，风险预警

贷后管理：优良客户复借，流失客户捞取，失信客户催收


从申请人的角度出发，整个信贷申请的流程大概要经历这些阶段：

1、申请人发起申请；

2、进入反欺诈体系进行初审，有的公司会使用F卡（反欺诈评分卡），或者反欺诈规则包括内部和外部的黑名单、法院记录等，如果用户命中反欺诈规则或评分过低则直接拒绝；

3、通过反欺诈体系之后，进入我们的信审环节，这一步主要是通过人工的方式对用户的资料进行审核，例如用户资料是否完整、用户的年龄是否符合法律规定等，需要注意的是，不同公司的不同信贷产品可能都存在一些规则，例如不允许年龄小于18和大于60岁的人申请，不允许年收入低于XX元的人进行申请等等，当然，这部分排除规则的工作有时候也会由算法工程师来完成；

4、用户经历了上述的两个环节之后，就正式进入了A卡的体系，通过申请评分卡对用户的风险进行评级，低于XX分的用户进行拒绝，否则信用评级通过并将通过用户的名单提交给业务进行进一步的审批，关于业务进一步是如何审查可见：https://www.sohu.com/a/230996733_499067

5、业务审批通过，根据用户的申请评分结果，进行额度管理和利率定价（低分额度低利率高，高分额度高利率低）此时用户正式进入了贷后跟踪环节；

6、贷后管理过程中就需要我们使用到B卡，即用户的行为评分卡，根据用户放款后的行为动态对用户的风险进行评估从而进行额度调整、利率调整，除此之外，对于老户的营销建模很多时候也是在这一步进行的，但是为了理解上的便于区分并且实际上营销建模也会涉及到新户方面的获客，我们将营销建模单独作为一个独立的流程后续进行阐述；

7、最后就到了C卡的环节，C卡的主要目的在于降低入催率（更少的人逾期），提高出催率（更多的逾期账单被结清），C卡分为失联模型、还款率模型、迁移率模型，具体的迁移率模型预测用户的逾期状态变坏的概率、还款率模型用于预测对逾期用户进行催收最终能催收回多少、失联模型预测逾期用户是否会发生失联；

8、当催收失败之后，对于呆账就会进行委外、法律诉讼等借助外部力量来尽量催回，将这些滞账进行打包销售等

上述描述了标准的信贷的整个流程，当然并不是每家机构都会严格按照上述的流程进行，例如C卡，很多公司就没有这类建模项目而较多选择委外的方式进行催回等；

# 二、业务详细描述

## 1.贷前反欺诈

从信贷欺诈的人数上一般分为个人欺诈和团伙欺诈，其中以团伙欺诈的危害最大，而从反欺诈的欺诈方来说目前较为常见的有第一方欺诈和第三方欺诈，一般来说第一方欺诈即从申请者的角度出发，客户本人申请贷款，恶意拖欠，或者是一人多身份证贷款意愿不良的，通过活体检测和设备指纹技术可以拦截大部分风险，目前信贷领域的反欺诈常常针对于第三方欺诈，即从独立于客户与公司的独立机构的角度出发，例如黑产获取客户的敏感资料，冒充客户身份申请贷款等；关于反欺诈的问题，下文会详细描述；




信贷业务中的反欺诈只是反欺诈领域的一个小分支，实际上留意过招聘市场的同学应该知道，很多公司都有招聘各类的反欺诈工程师，从互联网金融业务到电商线上的业务等等，反欺诈是一个非常庞大的领域，其中应用的技术是非常的全面，因为反欺诈的业务场景和数据场景都各不相同，例如贷前反欺诈，活体检测、身份证验证等就涉及到CV领域的算法，电商线上促销的薅羊毛问题，常常涉及到设备的社交网络关系，就需要用到图算法领域的知识等等；

## 2.评分卡相关

授信评分：
     
     基本属性：学历，工作，手机号，身份证 （手机号归属地查询，身份证获取性别年龄信息等）
     设备信息：手机品牌，IMEI，通讯录（app list特征衍生，app2vec,item2vec,设备指数技术（也常用于反欺诈场景等））
     人行征信：信用记录，资产（征信报告的时间窗衍生等）
     外部资信：芝麻信用，百融，同盾（芝麻信用，百融，同盾等，常进行外部数据探索（买数据要钱！））
     关系网络：亲密联系人，通话记录数（亲密联系人，通话记录数，构建知识图谱，图特征提取，graph embedding,GCN等）
     行为信息：App偏好，活动（序列数据的时间窗衍生，RNN\CNN\Transformer embedding）

### 2.1 A卡介绍

这里略过项目可行性分析、数据探察等前期繁琐的工作直接介绍A卡的构建流程

1、滚动率分析确定好坏客户定义；

2、vintage分析确定用户的“好坏”的表现期长度；

3、根据政策、业务规则、用户开户是否使用等设定用户排除规则，将无法建模的用户剔除；

4、外部数据探索（如有）

5、特征工程、特征稳定性、特征的业务可解释性等

6、模型构建与调优

7、模型评估、跨时间样本验证、模型报告等；

8、模型上线与监控；

9、随时准备面对和解决线上可能出现的各种问题，例如模型衰减、线上线下特征工程逻辑不一致、任务调度出现的各种诡异bug等以及其它的一些延申的项目的实施例如根据不同的客群建立子模型等；

相对于B、C卡，A卡的重要程度最高，因为贷前风控往往是整个风控项目的重中之重，贷前做好了，贷中和贷后就相对容易一些；


### 2.2 B卡介绍

B卡相对于A卡，引入了更多用户贷后的行为信息，因此一般模型的表现要强于A卡；

1、用户行为信息包括消费行为（信用卡账单、银行流水等、用户过去的逾期行为、用户登陆app的行为）的收集与处理；

2、结合A卡的特征训练模型；

3、后续流程和A卡大致相同不赘述

根据借款人放贷后的行为表现，预测未来逾期风险，引入B卡之后我们可以动态监控用户的违约风险，并根据用户的违约风险进行额度管理、利率调整等


### 2.3 C卡介绍

C卡相对于A、B卡，引入了更多用户贷后的还款意愿（申请贷款行为、履约记录、催收记录（如果为老客户，新客户首次入催则没有））等信息；

C卡大体上有三种模型的构建：

1、迁移率模型——预测逾期人群从轻度逾期变成重度逾期的概率；

2、还款率预测模型——预测经过催收后，最终催收回的欠款比例；

3、失联模型——对于尚未能联系上的用户预测其未来失联的概率；

催收评分卡相对于A、B卡来说出场的几率较低，因为很多信贷公司常常选择进行委外处理，对于C卡来说，其目标是降低入催率（更少的人逾期），提高出催率（更多的逾期账单被结清）。
